```python
# '''ETL for consumer complaints data table'''
%reset -f
from time import process_time 
import sys
from complaints_fxns import open_and_read_csv, csv_data_to_dict, generate_results_list, greater_than_one, write_results_to_csv#, csv_data_to_dict_alt

# command line arguments
# filename, csv_input, csv_output = sys.argv
# csv_input = '../input/complaints.csv' 
# csv_output = '../output/report.csv'
csv_input = '../insight_testsuite/mytest_2/input/test_data.csv' 
csv_output = '../insight_testsuite/mytest_2/output/report.csv'
```

```python
import csv
from collections import Counter
import operator

tic = process_time()

## EXTRACT ##
print('extract')

# create dictionary from csv data
# def open_and_read_csv(filepath):
# '''docstring for open_and_read_csv
# input: file and path to csv data
# output: list of OrderedDict objects'''
filepath = csv_input
my_data=[]
with open(filepath) as csvfile:
    reader = csv.DictReader(csvfile)
#     my_data = [row for row in reader]
    for row in reader:
        my_data.append(row)
# return my_data

# display(my_data)

toc = process_time()
print(toc - tic)
tic = process_time()


# d = csv_data_to_dict_alt(my_data)
# def csv_data_to_dict_alt(my_data):
# '''docstring for csv_data_to_dict
# input: list of OrderedDict objects
# output: return dictionary of key:value (column header: list of elements)'''

# construct list of lists
l = []
keys = list(dict(my_data[0]))
l = [[dict(my_data[0])[i]] for i in keys]
# for i in keys:
#     l.append([dict(my_data[0])[i]])

# extract remaining data into list of lists
for x in my_data[1:]:
    y = dict(x)
#     l = [l[i].append(y[keys[i]]) for i in range(len(keys))]
    for i in range(len(keys)):
        l[i].append(y[keys[i]])

# convert list of lists to dictionary
d = {}
d = {i: j for (i, j) in zip(keys,l)}
# for i,j in zip(keys,l):
#     d[i] = j

# return d

toc = process_time()
print(toc - tic)

d
```

```python
tic = process_time()

## TRANSFORM ##
print('transform')
# selection of columns for analysis
col_prod = 'Product'
col_date = 'Date received'
# loop through product and year instances and calculate complaint and company metrics
results = []
r = []
# product loop
for p in list(set(d[col_prod])):
    # positions
    pos = [i for i, e in enumerate(d[col_prod]) if e == p]
    # corresponding dates
    dates = [d[col_date][i].split('-')[0] for i in pos]
    # date loop
    for x in list(set(dates)):
        results.append(generate_results_list(d,p,x,dates))
        
toc = process_time()
toc - tic
```

```python
tic = process_time()

## LOAD ##
print('load')
# field names
fields = ['product','year','complaint_count','company_count*','complaint_percent_by_worst_company']
# write file
write_results_to_csv(fields,csv_output,results)

toc = process_time()
toc - tic
```
